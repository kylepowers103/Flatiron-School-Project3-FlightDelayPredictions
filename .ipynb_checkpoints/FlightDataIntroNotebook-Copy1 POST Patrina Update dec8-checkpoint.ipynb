```python
#Get Dataframes of all 2017 Months and merge them 
files=["On_time2015_00.csv","On_time2015_01.csv","On_time2015_02.csv","On_time2015_03.csv","On_time2015_04.csv",'On_time2016_00.csv','On_time2016_01.csv','On_time2016_02.csv','On_time2016_03.csv','On_time2016_04.csv','On_time2017_00.csv','On_time2017_01.csv','On_time2017_02.csv','On_time2017_03.csv','On_time2017_04.csv']
file_total=[]
for file in files:
    new=pd.read_csv('compressed/'+file)#,usecols=['Month', 'DayofMonth', 'DayOfWeek','Origin', 'Dest', 'UniqueCarrier', 'FlightNum', 'Origin', 'Dest', 'DepTime', 'DepDelay', 'TaxiOut', 'WheelsOff', 'WheelsOn', 'TaxiIn', 'ArrTime', 'ArrDelay', 'Cancelled', 'Diverted', 'AirTime', 'Flights', 'Distance', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay', 'DivAirportLandings'])
    file_total.append(new)
    
merged_df = pd.concat(file_total, axis = 0)
merged_df.describe()
```

```python
#merge categories for all columns used 
df=merged_df[['CRSDepTime', 'DepDelayMinutes',"Reporting_Airline","Flight_Number_Reporting_Airline", 'DepDel15', 'DepartureDelayGroups','Month', 'DayofMonth', 'Origin', 'Dest', 'DepTime', 'DepDelay', 'TaxiOut', 'TaxiIn', 'ArrTime', 'ArrDelay', 'Cancelled', 'Diverted', 'AirTime', 'Flights', 'Distance', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay', 'DivAirportLandings',"CancellationCode"]]
subset_df = df[(df['Origin'] == 'ATL') | (df['Origin'] == 'ORD') | (df['Origin'] == 'DFW') | (df['Origin'] == 'CLT') | (df['Origin'] == 'DEN')]
```

```python
#Get Flight Number as a combination of Carrier code and individual flight number. E.g 'AA' and '346' becom 'AA346'
subset_df['FlightNum'] = subset_df['Reporting_Airline'] + subset_df['Flight_Number_Reporting_Airline'].astype('str')
#drop old flightnumber
subset_df.drop(columns = 'Flight_Number_Reporting_Airline', inplace = True)
subset_df.drop(columns = 'DivAirportLandings', inplace = True)
```
